```python
a = list(df.Glucose)
print(a)
```

```python
b = list(df.BloodPressure)
print(b)
```

```python
c = list(df.Outcome)
print(c)
count = len(c)
```

```python
nx=148
ny=72
#ny=float(y)
ans=[]
e=[]
for i in range(0, count):
  ans.append(math.sqrt(((nx-a[i])*(nx-a[i]))+((ny-b[i])*(ny-b[i]))))
print(ans)
#d=ans.index(min(ans))
#print('Index=',d)
#print('class label=',c[d])
#cd=c[d]
e=sorted(ans)
k=int(input("Enter k value:"))
cc=[]

for x in range(0,k):
  sd=ans.index(e[x])
  cc.append(c[sd])

print("K list:",cc)
```

```python
def most_frequent(List):
    return max(set(List), key = List.count)

print("Class label :",most_frequent(cc))
```
